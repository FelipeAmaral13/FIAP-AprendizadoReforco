# Exemplo 2 - Frozen Lake

Este ambiente  ilustrado na figura abaixo.



Temos 16 estados, denotados por 0,1,2,,15.  Cada estado corresponde a um determinado campo que pode ser um campo congelado (F), buraco (H), estado inicial (S), do estado objetivo (G).  Nosso objetivo  atingir o estado objetivo comeando pelo estado inicial e evitando o estado hole.



Nesta clula, estamos importando as bibliotecas necessrias e criando o ambiente "FrozenLake-v1" com renderizao visual. A renderizao visual permitir que observemos a simulao do ambiente.

```python
# Importando as bibliotecas necessrias
import gym
import time

# Criando o ambiente FrozenLake-v1 com renderizao visual
env = gym.make("FrozenLake-v1", render_mode='human')
```

```python
# observation space - states 
print(env.observation_space)
 
# actions: left -0, down - 1, right - 2, up- 3
print(env.action_space)
```

Aqui, estamos reiniciando o ambiente para o estado inicial e exibindo visualmente o estado usando env.render(). Tambm estamos imprimindo uma mensagem para indicar que estamos no estado inicial do sistema.

```python
# Reiniciando o ambiente e exibindo o estado inicial
env.reset()
env.render()

print('Estado inicial do sistema')
```

```python
# Definindo o nmero de iteraes
numberOfIterations = 5
```

Nesta clula, estamos definindo o nmero de iteraes que desejamos executar no loop.

Aqui, estamos executando o loop com base no nmero de iteraes definido anteriormente. A cada iterao:

* Geramos uma ao aleatria do espao de ao do ambiente.
* Executamos a ao no ambiente usando env.step(randomAction) e obtemos informaes sobre o estado resultante.
* Exibimos o ambiente renderizado usando env.render() para visualizar a mudana no estado.
* Imprimimos informaes sobre a iterao, incluindo o nmero da iterao e a ao escolhida.
* Aguardamos por um curto perodo de tempo usando time.sleep(2) para que possamos observar as mudanas no ambiente.
* Verificamos se o episdio foi concludo usando returnValue[2] e, se for verdadeiro, interrompemos o loop.

```python
# Executando interaes com o ambiente
for i in range(numberOfIterations):
    # Gerando uma ao aleatria do espao de ao
    randomAction = env.action_space.sample()

    # Executando a ao no ambiente e obtendo informaes
    observation, reward, terminated, truncated, info = env.step(randomAction)

    # Exibindo o ambiente renderizado
    env.render()

    # Exibindo informaes sobre a iterao
    print(f'Iterao: {i+1}, ao {randomAction} resultou em: {observation, reward, terminated, truncated, info} ')

    # Aguardando por um curto perodo de tempo
    time.sleep(2)

    # Verificando se o episdio foi concludo
    if terminated:
        break
```

Por exemplo:

`(0, 0.0, False, False, {prob: 0.3333333333333333})`

Isso implica que o estado final  0, a recompensa  0, False significa que o estado final no  um estado terminal (no chegamos ao buraco ou ao estado objetivo), outro valor de retorno booleano False no  importante para esta discusso, e {'prob': 0,3333333333333333}  a probabilidade de transio de nosso estado inicial 0 para o estado 0.

```python
# Fechando o ambiente
env.close()
```
